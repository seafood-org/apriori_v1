```python
import matplotlib.pyplot as plt
import numpy as np

# Table data
algorithms = ['apriori naïve', 'apriori optmized', 'fptree']
worker_nums = [2, 3, 4, 5]
runtimes = [[558.93, 464.43, 505.33, 428.25],
           [519.41, 294.00, 245.96, 218.93],
           [24.43, 24.97, 30.14, 29.17]]

# Create the plot
plt.figure(figsize=(10, 6))

for i, algorithm in enumerate(algorithms):
    plt.plot(worker_nums, runtimes[i], marker='o', label=algorithm)

plt.xlabel('Number of Workers')
plt.ylabel('Runtime (s)')
plt.title('Algorithm Performance with Varying Worker Numbers')
plt.xticks(worker_nums)
plt.legend()
plt.grid()
plt.show()
```
